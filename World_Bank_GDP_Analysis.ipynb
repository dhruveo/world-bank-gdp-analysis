```python
import pandas as pd
df = pd.read_csv("india_gdp.csv", header=None)
df.columns = ['Year', 'GDP']
df.head()
```

```python
df.describe()
df.isnull().sum()
df.info()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['Year'], df['GDP'], marker='o')
plt.title('India GDP Over Time')
plt.xlabel('Year')
plt.ylabel('GDP (USD)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
top10 = df.nlargest(10, 'GDP')
plt.figure(figsize=(10,5))
plt.bar(top10['Year'].astype(str), top10['GDP'])
plt.title("Top 10 Years with Highest GDP")
plt.xlabel("Year")
plt.ylabel("GDP (USD)")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['GDP'], bins=10, edgecolor='black')
plt.title("Distribution of India's GDP")
plt.xlabel("GDP (USD)")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(5,7))
plt.boxplot(df['GDP'])
plt.title("Box Plot of GDP")
plt.ylabel("GDP (USD)")
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.scatter(df['Year'], df['GDP'])

plt.title("Scatter Plot of GDP vs Year")
plt.xlabel("Year")
plt.ylabel("GDP (USD)")

plt.grid(True)

plt.show()
```

```python
df = df.sort_values("Year")

df["Moving_Average"] = df["GDP"].rolling(window=5).mean()

plt.figure(figsize=(10,5))

plt.plot(df["Year"], df["GDP"], label="GDP")
plt.plot(df["Year"], df["Moving_Average"], linewidth=3, label="5-Year Moving Average")

plt.title("GDP and 5-Year Moving Average")
plt.xlabel("Year")
plt.ylabel("GDP (USD)")
plt.legend()
plt.grid(True)

plt.show()
```
